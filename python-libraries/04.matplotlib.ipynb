```python
a = np.array([2,3,4,5,1,6])
b = np.array([3,4,5,2,1,7])

c = np.array([6,8,3,7,4,2])
d = np.array([2,3,4,5,1,6])

plt.subplot(1, 2, 1)
plt.plot(a,b)
plt.title('First subplot')

plt.subplot(1, 2, 2)
plt.plot(c,d)
plt.title('Second subplot')

plt.suptitle("Two plots")

plt.show()
```
